```python
import cv2 as cv
import numpy as np
import glob
import os
```

```python
class Calibarator:
    '''
        标定类
    '''
    def __init__(self, square_length: int, width_corners:int, height_corners: int):
        self.square_length = square_length # set length of each square
        self.width = width_corners # 宽方向的内角点个数
        self.height = height_corners # 长方向的内焦点个数
        
        # 单目标定参数
        self.images = [] # save loaded images
        self.pixel_points = [] # save 2d points
        self.real_points = [] # save 3d points 
        self.r_vecs = [] # save rotation vectors
        self.t_vecs = [] # save translate vectors
        self.cameraMatrix = np.zeros([3,3],dtype = np.float32) # 相机矩阵
        self.distCoeffs = np.zeros([1,5],dtype = np.float32) # 畸变系数
        
        # 双目标定参数
        self.left_images = [] # 存左侧图片
        self.right_images = [] # 存右侧图片
        self.object_points = [] # 物点
        self.left_images_points = [] # 左像点
        self.right_images_points = [] # 右像点
        self.left_camera_matrix = np.zeros([3, 3], dtype=np.float32) # 左相机矩阵
        self.right_camera_matrix = np.zeros([3, 3], dtype=np.float32) # 右相机矩阵
        self.left_dist_coeffs = np.zeros([1, 5], dtype=np.float32) # 左畸变系数
        self.right_dist_coeffs = np.zeros([1, 5], dtype=np.float32) # 右畸变系数
        
        # 双目相机校正
        self.R = [] # 旋转矩阵
        self.T = [] # 平移
        self.E = [] # 本质矩阵
        self.F = [] # 基本矩阵
        self.R1 = [] # 左相机旋转矩阵
        self.R2 = [] # 右相机旋转矩阵
        self.T1 = [] # 左相机平移矩阵
        self.T2 = [] # 右相机旋转矩阵
        self.Q = [] 
        self.map_left_x = [] # 左侧相机x映射
        self.map_left_y = [] # 左侧相机y映射
        self.map_right_x = [] # 右侧相机x映射
        self.map_right_y = [] # 右侧相机y映射
        
        
        
        for i in range(1, 15):
            if i != 10:
                img = cv.imread("./left/left{0}.jpg".format(str(i).zfill(2)))
                self.images.append(img)
            
    def calc_image_points(self, show=False):
        ob_p = []
        for i in range(0, self.height):
            for j in range(0, self.width):
                ob_p.append(np.array([j * self.square_length, i * self.square_length, 0], dtype=np.float32)) 
        
        
        for img in self.images:
            # 找到棋盘格内角点
            self.real_points.append(ob_p)
            pattern_found, im_p = cv.findChessboardCorners(img, (self.width, self.height), cv.CALIB_CB_ADAPTIVE_THRESH+ cv.CALIB_CB_FAST_CHECK + cv.CALIB_CB_NORMALIZE_IMAGE)
            if(pattern_found):
                gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
                cv.cornerSubPix(gray, im_p, (5,5), (-1,-1), (cv.TERM_CRITERIA_EPS | cv.TERM_CRITERIA_MAX_ITER, 30, 0.1))
                self.pixel_points.append(im_p)
                if(show):
                    img_to_draw = np.array(img)
                    cv.drawChessboardCorners(img_to_draw, (self.width, self.height), im_p, False) # 最后一个false表示不画它们的连线
                    cv.imshow("Chessboard corners", img_to_draw)
                    cv.waitKey(0)
                    cv.destroyAllWindows()
                    # fix the bug that on mac destroyWindows do not work normally.
                    cv.waitKey(1)
                    cv.waitKey(1)
                    cv.waitKey(1)
                    cv.waitKey(1)
                
            else:
                raise Exception("没找到！")
          
    def calibrate(self):
        result = cv.calibrateCamera(np.float32(self.real_points), np.float32(self.pixel_points), (self.images[0].shape[1], self.images[0].shape[0]),self.cameraMatrix, self.distCoeffs, self.r_vecs, self.t_vecs, 0)
        self.rms_error = result[0]
        self.cameraMatrix = result[1]
        self.distCoeffs = result[2]
        self.r_vecs = result[3]
        self.t_vecs = result[4]
        print("误差:",self.rms_error)
        print("相机矩阵\n",self.cameraMatrix)
        print("畸变系数\n",self.distCoeffs)
        print("旋转向量\n",self.r_vecs)
        print("平移向量\n",self.t_vecs)
        
    def correct_image(self):
        i = 0
        if not os.path.exists("./left_correct"):
            os.mkdir("./left_correct")
        for img in self.images:
            dst = cv.undistort(img, self.cameraMatrix, self.distCoeffs)
            cv.imwrite("./left_correct/{0}_corrected.jpg".format(i), dst)
            i += 1
        print("修正结果保存在./left_correct目录")
            
            
    def stereo_calc_images_points(self, show=False):
        ob_p = []
        for i in range(0, self.height):
            for j in range(0, self.width):
                ob_p.append(np.array([j * self.square_length, i * self.square_length, 0], dtype=np.float32)) 
        
        # 左侧像点
        for i in range(1, 15):
            if i != 10:
                img = cv.imread("./left/left{0}.jpg".format(str(i).zfill(2)))
                self.left_images.append(img)
        
        for img in self.left_images:
            self.object_points.append(ob_p)
            pattern_found, im_p = cv.findChessboardCorners(img, (self.width, self.height), cv.CALIB_CB_ADAPTIVE_THRESH+ cv.CALIB_CB_FAST_CHECK + cv.CALIB_CB_NORMALIZE_IMAGE)
            if(pattern_found):
                gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
                cv.cornerSubPix(gray, im_p, (5,5), (-1,-1), (cv.TERM_CRITERIA_EPS | cv.TERM_CRITERIA_MAX_ITER, 30, 0.1))
                self.left_images_points.append(im_p)
                if(show):
                    img_to_draw = np.array(img)
                    cv.drawChessboardCorners(img_to_draw, (self.width, self.height), im_p, False) # 最后一个false表示不画它们的连线
                    cv.imshow("Chessboard corners", img_to_draw)
                    cv.waitKey(0)
                    cv.destroyAllWindows()
                    # fix the bug that on mac destroyWindows do not work normally.
                    cv.waitKey(1)
                    cv.waitKey(1)
                    cv.waitKey(1)
                    cv.waitKey(1)
                
            else:
                raise Exception("没找到！")
            
        
        # 右侧像点
        for i in range(1, 15):
            if i != 10:
                img = cv.imread("./right/right{0}.jpg".format(str(i).zfill(2)))
                self.right_images.append(img)
                
        
        for img in self.right_images:
            pattern_found, im_p = cv.findChessboardCorners(img, (self.width, self.height), cv.CALIB_CB_ADAPTIVE_THRESH+ cv.CALIB_CB_FAST_CHECK + cv.CALIB_CB_NORMALIZE_IMAGE)
            if(pattern_found):
                gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
                cv.cornerSubPix(gray, im_p, (5,5), (-1,-1), (cv.TERM_CRITERIA_EPS | cv.TERM_CRITERIA_MAX_ITER, 30, 0.1))
                self.right_images_points.append(im_p)
                if(show):
                    img_to_draw = np.array(img)
                    cv.drawChessboardCorners(img_to_draw, (self.width, self.height), im_p, False) # 最后一个false表示不画它们的连线
                    cv.imshow("Chessboard corners", img_to_draw)
                    cv.waitKey(0)
                    cv.destroyAllWindows()
                    # fix the bug that on mac destroyWindows do not work normally.
                    cv.waitKey(1)
                    cv.waitKey(1)
                    cv.waitKey(1)
                    cv.waitKey(1)
                
            else:
                raise Exception("没找到！")
            
            
    def stereo_calibrate(self):
        lresult = cv.calibrateCamera(np.float32(self.object_points), self.left_images_points, (self.left_images[0].shape[1], self.left_images[0].shape[0]), self.left_camera_matrix, self.left_dist_coeffs)
        self.left_camera_matrix = lresult[1]
        self.left_dist_coeffs = lresult[2]
        rresult = cv.calibrateCamera(np.float32(self.object_points), self.right_images_points, (self.right_images[0].shape[1], self.right_images[0].shape[0]), self.right_camera_matrix, self.right_dist_coeffs)
        self.right_camera_matrix = rresult[1]
        self.right_dist_coeffs = rresult[2]
        result = cv.stereoCalibrate(np.float32(self.object_points), np.float32(self.left_images_points), np.float32(self.right_images_points), self.left_camera_matrix, self.left_dist_coeffs,
                                   self.right_camera_matrix, self.right_dist_coeffs, (self.images[0].shape[1], self.images[0].shape[0]))
        
        self.R = result[5]
        self.T = result[6]
        self.E = result[7]
        self.F = result[8]
        print("retval \n", result[0])
        print("left camera matrix\n", result[1])
        print("left camera distCoeffs\n", result[2])
        print("right camera matrix\n", result[3])
        print("right camera distCoeffs\n", result[4])
        print("R 旋转\n", result[5])
        print("T 平移\n", result[6])
        print("E 本质\n", result[7])
        print("F 基础\n", result[8])
    
    def stereo_rectify(self):
        result = cv.stereoRectify(self.left_camera_matrix, self.left_dist_coeffs, 
                         self.right_camera_matrix, self.right_dist_coeffs, (self.images[0].shape[1], self.images[0].shape[0]), self.R, self.T)
        self.R1 = result[0]
        self.R2 = result[1]
        self.T1 = result[2]
        self.T2 = result[3]
        self.Q = result[4]
        print("Result\n", result)
        print("R left\n", self.R1)
        print("R right\n", self.R2)
        print("T left\n", self.T1)
        print("T right\n", self.T2)
        print("Q \n",self.Q)
        
        # 计算映射
        self.map_left_x, self.map_left_y = cv.initUndistortRectifyMap(self.left_camera_matrix, self.left_dist_coeffs, self.R1, self.T1, (self.images[0].shape[1],self.images[0].shape[0]), cv.CV_32FC1)
        self.map_right_x, self.map_right_y = cv.initUndistortRectifyMap(self.right_camera_matrix, self.right_dist_coeffs, self.R2, self.T2, (self.images[0].shape[1],self.images[0].shape[0]), cv.CV_32FC1)
        
        if not os.path.exists("./rectify"):
            os.mkdir("./rectify")
        # 恢复
        for i in range(len(self.left_images)):
            dst_left = cv.remap(self.left_images[i], self.map_left_x, self.map_left_y, cv.INTER_LINEAR)
            # 画出极线
            epipolar_line_left = cv.computeCorrespondEpilines(np.float32(np.array([[100, 100], [130, 130], [175, 175]])), 1, self.F)
            # print("左侧相机极线\t", epipolar_line_left)
            for line in epipolar_line_left:
                cv.line(dst_left, (0, int(-line[0][2]/line[0][1])), (640, int(-(line[0][2]+line[0][0]*640)/line[0][1])), (0, 255, 0))
            # 保存图片
            cv.imwrite("./rectify/left_{0}.jpg".format(i), dst_left)
            dst_right = cv.remap(self.right_images[i], self.map_right_x, self.map_right_y, cv.INTER_LINEAR)
            epipolar_line_right = cv.computeCorrespondEpilines(np.float32(np.array([[100, 100], [130, 130], [175, 175]])), 2, self.F)
            # print("右侧相机极线\t", epipolar_line_right)
            for line in epipolar_line_right:
                cv.line(dst_right, (0, int(-line[0][2]/line[0][1])), (640, int(-(line[0][2]+line[0][0]*640)/line[0][1])), (0, 255, 0))
            cv.imwrite("./rectify/right_{0}.jpg".format(i), dst_right)
        
        
        print("修正结果存储在./rectify路径下")
        
    def calc_disparity(self, min_disp=16, num_disp=96, blockSize = 16, window_size=3, disp12MaxDiff=1, uniquenessRatio=10, whichImg=0, ifGray=True):
        """
        min_disp
        num_disp: 必须为16倍数
        whichImg: 0 ~ 13
        """
        stereo = cv.StereoSGBM_create(min_disp, num_disp, blockSize, P1 = 8*3*window_size**2,
        P2 = 32*3*window_size**2,
        disp12MaxDiff = 1,
        uniquenessRatio = 10,
        speckleWindowSize = 100,
        speckleRange = 32);
        dst_left = cv.remap(self.left_images[whichImg], self.map_left_x, self.map_left_y, cv.INTER_LINEAR)
        dst_right = cv.remap(self.right_images[whichImg], self.map_right_x, self.map_right_y, cv.INTER_LINEAR)
        if ifGray:
            grayL = cv.cvtColor(dst_left, cv.COLOR_BGR2GRAY)
            grayR = cv.cvtColor(dst_right,  cv.COLOR_BGR2GRAY)
            disp = stereo.compute(grayL, grayR).astype(np.float32)
        else:
            disp = stereo.compute(dst_left, dst_right).astype(np.float32)
        cv.imwrite("Disparity_{0}.jpg".format(whichImg), (disp-min_disp)/num_disp)
        cv.imwrite("Disparity_{0}_ori.jpg".format(whichImg), disp/16)
        
        
```

```python
calib = Calibarator(25.,9, 6) # 初始化标定器
# 单目标定
calib.calc_image_points() # *计算内角点

calib.calibrate()
calib.correct_image()
```

```python
# 双目标定
calib.stereo_calc_images_points()
calib.stereo_calibrate()
```
